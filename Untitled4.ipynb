```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la página con los datos de Tesla (ejemplo con Yahoo Finance)
url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'

# Obtener la página web
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Buscar la tabla de datos de ingresos
# Este es un ejemplo de cómo encontrar la tabla (ajusta el selector si es necesario)
table = soup.find('table', {'class': 'table'})

# Extraer las filas de la tabla
rows = table.find_all('tr')

# Crear una lista para almacenar los datos
data = []

# Iterar sobre las filas para extraer las celdas
for row in rows[1:]:  # Skip the header row
    cells = row.find_all('td')
    if len(cells) > 1:
        year = cells[0].get_text(strip=True)
        revenue = cells[1].get_text(strip=True).replace('$', '').replace(',', '')
        data.append([year, float(revenue)])

# Crear un DataFrame con los datos de ingresos
tesla_revenue = pd.DataFrame(data, columns=['Year', 'Revenue'])

# Mostrar las últimas 5 filas
print(tesla_revenue.tail())
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Define the URL for the financial data of GameStop (GME)
url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'

# Send an HTTP request to the website and get the page content
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the revenue data table (may need to adjust based on the website structure)
table = soup.find('table', {'class': 'historical_data_table'})

# Extract the rows from the table
rows = table.find_all('tr')

# Create an empty list to store the data
data = []

# Loop through each row and extract the columns
for row in rows[1:]:  # Skip the header row
    cols = row.find_all('td')
    cols = [col.text.strip() for col in cols]
    data.append(cols)

# Create a DataFrame from the scraped data
columns = ['Year', 'Revenue']
gme_revenue = pd.DataFrame(data, columns=columns)

# Display the last five rows
print(gme_revenue.tail())
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch Tesla stock data from Yahoo Finance
tesla_data = yf.download('TSLA', start='2020-01-01', end='2024-01-01')

# Define the function to plot the stock data
def make_graph(data):
    # Plot the 'Close' price of Tesla stock
    plt.figure(figsize=(10, 6))
    plt.plot(data['Close'], label='Tesla Close Price', color='blue')
    plt.title('Tesla Stock Price (2020-2024)')
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function to plot the data
make_graph(tesla_data)
```
